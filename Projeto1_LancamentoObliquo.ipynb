#Descrição Geral do Sistema Simulado

A simulação apresenta o trajeto, velocidades e a energia de um projétil, que faz um movimento oblíquo até encostar no solo. Em nossos cálculos, levaremos em conta que a gravidade (G) terá sentido negativo (-).

 Modelo matemático

*   Velocidade Vertical:

  $
  ΔVy = (G-\frac{K}{m}*v)*\Delta t
  $

*   Velocidade Horizontal:

  $
  \Delta Vx = (\frac{-K*V0*cos(\Theta)}{m})*\Delta t
  $

*   Deslocamento Vertical:

  $
  \Delta Dx = Vy*\Delta t
  $

*   Deslocamento Horizontal:

  $
  \Delta Dx = Vx*\Delta t
  $

*   Energia Total:

  $
  Et = Ep+Ec
  $

*   Energia Cinética:

  $
  Ec = \frac{m*Vx}{2}
  $

*   Energia Potencial Gravitacional:

  $
  Ep=m*g*h
  $

# Parâmetros da Simulação

### Simulação 01:

- $V0 = 30m/s$ - Velocidade inicial em que o objeto será lançado.
- $g = 10$ - Aceleração da gravidade que irá atuar sobre o corpo.
- $m = 10$ - Massa do corpo a ser lançado.
- $k = 1$ - Coeficiente de atrito levado em consideração. Mais especificamente, refere-se à resistência do ar.
- $\Theta = \pi/4$ - Ângulo no qual o lançamento será realizado, em radianos.
- $(x, y) = $(0, 0) - Altura inicial do corpo, em coordenadas x e y.

### Simulação 02:

- $(x, y) = $(5, 30) - Altura inicial do corpo, em coordenadas x e y.

### Simulação 03:

- $g = 5$ - Aceleração da gravidade que irá atuar sobre o corpo.

- $(x, y) = $(0, 10) - Altura inicial do corpo, em coordenadas x e y.


```python
# Bibliotecas
import math
import numpy as np
from matplotlib import pyplot as plt
```

#Simulação 1 - Cálculos e Resultados

#Parâmetros

```python
DT = 0.01                 # Variação do tempo
Y0 = 0                    # Altura inicial
X0 = 0                    # Posição inicial
V0 = 30                   # Velocidade inicial
A = math.pi/4             # Angulo do lancamento em radiano
Vx = V0*(math.cos(A))     # Velocidade em X
Vy = V0*(math.sin(A))     # Velocidade em Y
G = -10                   # Gravidade
TMax = 10                 # Tempo máximo de simulação
K = 1                     # Constante de atrito
m = 10                    # Massa em kg

# Variáveis
vx = Vx
vy = Vy 
y = Y0
x = X0 

# Listas para plot
vxl = []                      # Velocidade em x
vyl = []                      # Velocidade em y
sxl = []                      # Deslocamento em x
syl = []                      # Deslocamento em y
epl = []                      # Energia potencial gravitacional
ecl = []                      # Energia cinetica
etl = []                      # Energia total = Ep+Ec
tl = np.arange(DT, TMax, DT)  # Tempo
```

#Cálculos

```python
# Velocidades Instantâneas
vxi = vx
vyi = vy

# Energias
ep = 0
ec = 0
et = 0

sx = x  # deslocamento inicial em x
sy = y  # deslocamento inicial em y

for i in np.arange(DT, TMax, DT):
  vxi = vxi + (-1) * ((K * vxi ) / m ) * DT
  vyi = vyi + G * DT - ((K * vyi) / m) * DT
    
  DSx = vxi * DT
  DSy = vyi * DT

  sx = sx + DSx
  sy = sy + DSy

  ep = -1 * m * G * sy
  ec = m * (vxi**2)/2
  et = ep + ec

  # Projétil toca o solo
  if sy <= 0:
    sy = 0

  vxl.append(vxi)
  vyl.append(vyi)

  sxl.append(sx)
  syl.append(sy)

  epl.append(ep)
  ecl.append(ec)
  etl.append(et)
```

#Gráficos e Análises

```python
# Gráfico 1 - velocidade no eixo X em função do tempo
plt.plot(tl, vxl)
plt.show()
```

```python
# Gráfico 2 - velocidade no eixo Y em função do tempo
plt.plot(tl, vyl)
plt.show()
```

```python
# Gráfico 3 - deslocamento no eixo X em função do tempo
plt.plot(tl, sxl)
plt.show()
```

```python
# Gráfico 4 - deslocamento no eixo Y em função do tempo
plt.plot(tl, syl)
plt.show()
```

```python
# Gráfico 5 - deslocamento no eixo Y em função do tempo
plt.plot(sxl, syl)
plt.show()
```

```python
# Graficos 6, 7 e 8 - Energias Potencial, Cinética e Total
plt.title("Energia Potencial")
plt.plot(tl, epl)
plt.show()

plt.title("Energia Cinética")
plt.plot(tl, ecl)
plt.show()

plt.title("Energia Total/Mecânica")
plt.plot(tl, etl)
plt.show()
```

Analisando os gráficos 1 e 2, temos, em cada um deles, a representação das velocidades nos eixos X e Y do projétil lançado. No primeiro gráfico, que diz respeito à velocidade no eixo X do lançamento, observa-se que tal velocidade decresce, já que o atrito do ar em tal eixo produz um efeito retardante. Já no segundo gráfico, temos que a velocidade no eixo Y cai até alcançar uma altura máxima, com Vy=0, e depois volta a aumentar (no sentido contrário, por isso chega a -50) até que atinja o solo.

Nos gráficos 3, 4 e 5, temos os deslocamentos realizados por nosso corpo em cada eixo e, no quinto, os deslocamentos um em função do outro. Como pode-se ver, o deslocamento no eixo X apenas aumenta de forma mais rápida, inicialmente, e depois aumenta de forma mais contínua, já que o atrito do ar atua desta forma. Quando ao deslocamento em Y, representado no gráfico 4, temos que o projétil alcança uma altura limite, de aproximadamente 20 metros, e depois decai até atingir o "chão", que indicamos em código como sendo quando o deslocamento em Y fosse igual a 0. No quinto gráfico, analisamos como seria um deslocamento em função do outro, levando ao que seria um deslocamento mais próximo ao "real", dadas os parâmetros estabelecidos nesta simulação.

Por fim, temos os gráficos 6, 7 e 8. Estes gráficos dizem respeito, respectivamente, as três energias atuantes no lançamento oblíquo: a energia potencial gravitacional, a energia cinética e a energia total (ou mecânica). No gráfico 6, temos que a energia potencial alcança seu pico na altura máxima do lançamento, algo em torno dos 2000, e perde seu valor com o decorrer do lançamento, atingindo o solo e mantendo-se em 0. Quanto a energia cinética, representada no gráfico 7, observamos que ela apenas perde seu valor com o passar do tempo (isso no eixo X). Para finalizar, temos o gráfico que mostra a energia total de nosso sistema, representado pela soma das energias obtidas com o passar do tempo. No começo, a energia total sobe, já que a energia potencial também cresce, já no fim, como a energia potencial já está em zero, a energia mecânica - ou total - segue igual a energia cinética, já que ficamos com Et = 0 + Ec.
