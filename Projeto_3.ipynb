*   Equações Diferenciais utilizadas nas simulações


  $
  \frac{dxP1}{dt} = (P1\cdot(p11-(p11\frac{P1}{K1})-p12\cdot H1-p13\cdot H2))
  $


  $
  \frac{dxP2}{dt} = (P2\cdot(p21-(p21 \frac{P2}{K2})-p22\cdot H2-p23\cdot H3))
  $


  $
  \frac{dxH1}{dt} = (H1\cdot(-p31+p32\cdot P1-p33\cdot C1))
  $


  $
  \frac{dxH2}{dt} = (H2\cdot(-p41+p42\cdot P2-p43\cdot C2))
  $


  $
  \frac{dxH3}{dt} = (H3\cdot (-p51+p52\cdot P1+p53\cdot P2-p54\cdot C2))
  $


  $
  \frac{dxC1}{dt} = (C1\cdot (-p61+p62\cdot H1-p63\cdot C3))
  $


  $
  \frac{dxC2}{dt} = (C2\cdot (-p71+p72\cdot H2+p73\cdot H3-p74\cdot C3))
  $


  $
  \frac{dxC3}{dt} = (C3\cdot (-p81+p82\cdot C1+p83\cdot C2))
  $



  ---
Cada equação representa uma espécie, levando a interação que eles possuem entre si, de predação e consumo.

#Parâmetros

```python
#Parametros
DT=0.01
Tmax=100

K1 = 3
K2 = 2

tl = np.arange(DT, Tmax, DT)

p11=0.6
p12=0.5
p13=0.5
p21=1
p22=1
p23=1
p31=0.01 
p32=1
p33=1
p41=0.29984
p42=1.9
p43=1
p51=0.8
p52=1.5
p53=2.0789
p54=1
p61=0.1
p62=1
p63=1
p71=0.01
p72=1.5
p73=1
p74=1
p81=1
p82=1
p83=1

#Populações iniciais

P1=1
P2=1
H1=1
H2=1
H3=1
C1=1
C2=1
C3=1
```

#Cálculos

```python
#Equacoes
P1l = []
P2l = []
H1l = []
H2l = []
H3l = []
C1l = []
C2l = []
C3l = []

for i in np.arange(DT, Tmax, DT):
    P1 += (P1 *(p11 - (p11 * (P1/K1)) - p12 * H1 - p13 *H2))* DT
    P2 += (P2 *(p21 - (p21 * (P2/K2)) - p22 * H2 - p23 *H3))* DT
    H1 += (H1 *(-p31 + p32 * P1 - p33 * C1))* DT
    H2 += (H2 *(-p41 + p42 * P2 - p43 * C2))* DT
    H3 += (H3 *(-p51 + p52 * P1 + p53 * P2 - p54 * C2))* DT
    C1 += (C1 *(-p61 + p62 * H1 - p63 * C3))* DT
    C2 += (C2 *(-p71 + p72 * H2 + p73 * H3 - p74 * C3))* DT
    C3 += (C3 *(-p81 + p82 * C1 + p83 * C2))* DT
    P1l.append(P1)
    P2l.append(P2)
    H1l.append(H1)
    H2l.append(H2)
    H3l.append(H3)
    C1l.append(C1)
    C2l.append(C2)
    C3l.append(C3)
```

#Gráficos e Análises

```python
#Plots

plt.plot(tl, P1l, color = "red")
plt.plot(tl, P2l, color =  "blue")
plt.plot(tl, H1l, color = "orange")
plt.plot(tl, H2l, color = "green")
plt.plot(tl, H3l, color = "yellow")
plt.plot(tl, C1l, color = "purple")
plt.plot(tl, C2l, color = "pink")
plt.plot(tl, C3l, color = "brown")
#ticks do plot
listOf_Yticks = np.linspace(0,4,20) 
plt.yticks(listOf_Yticks) 
plt.show()
```

Na primeira simulação, utilizamos todas as populações iguais, sendo todas 1. Com isso conforme as populações de herbivoros e carnivoros crescem bastante no inicio, enquanto as populações de produtores diminuem. Conforme o tempo avança, elas continuam oscilando, com tendencia de estabilizarem em um ponto fixo, neste caso, sem chegar a zero, matando uma população inteira.
